### Preprocessing the data

```python
#Getting runs based on each match

total_run_df=delivery.groupby(["match_id","inning"]).sum()["total_runs"].reset_index()
```

```python
total_run_df.head()
```

```python
# getting only first innings runs and storing it as target runs by adding one run to total run

total_run_df=total_run_df[total_run_df["inning"]==1]

total_run_df["total_runs"]+=1
```

```python
total_run_df.head()
```

```python
total_run_df.rename(columns={"total_runs":"target_runs"},inplace=True)
```

```python
total_run_df.drop(columns=["inning"],inplace=True)
total_run_df.head()
```

```python
match_df=match.merge(total_run_df,left_on="id",right_on='match_id')
match_df.head()
```

```python
match_df["team1"].unique()
```

```python
### There are few teams which are not playing this year such as Kochi, pune. so selecting only required teams.
teams = [
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Sunrisers Hyderabad',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
### Some teams changed their name after 2017, so correcting the team names

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')
```

```python
match_df=match_df[(match_df["team1"].isin(teams)) & (match_df["team2"].isin(teams)) ]
```

```python
match_df["team1"].unique()
```

```python
match_df.shape
```

```python
# we need matches where both teams played so delete rows where matches are decided based on dl method

match_df["dl_applied"].value_counts()
```

```python
match_df=match_df[match_df["dl_applied"]==0]
```

```python
match_df.columns
```

```python
# out of all columns, city, winner, terget_runs are most needed columns and match id to merge with delivery dataframe

match_df_final=match_df[["city","winner","target_runs","match_id"]]
```

```python
match_df_final.head()
```

```python
# merging match dataframe with delivery dataframe

delivery_df= match_df_final.merge(delivery,on="match_id")
delivery_df.head()
```

```python
### we need only total runs from 1st innings so we have that in target runs, we can eliminate those rows

delivery_df=delivery_df[delivery_df["inning"]==2]
```

I'm going to calculate probability for every ball in second innings, I dont have the data how much total runs the team is scoring in second inning at end of each ball. so i'm doing cumulative sum of runs

```python
delivery_df["current_score"]=delivery_df.groupby(["match_id"]).cumsum()["total_runs"]
```

```python
delivery_df.head()
```

```python
# Total runs left

delivery_df['runs_left'] = delivery_df['target_runs']-delivery_df['current_score']
delivery_df.head()
```

```python
# total balls left

delivery_df['balls_left'] = 120-((delivery_df['over']-1)*6+delivery_df['ball'])

delivery_df.head()
```

```python
''' Player dismissed column has nan and names, but we need only number of player dismissed in a match at every ball
so that we can have number of wickets in hand'''

delivery_df["player_dismissed"].fillna(0,inplace=True)
```

```python
delivery_df["player_dismissed"]=np.where(delivery_df["player_dismissed"]!=0,1,0)
```

```python
delivery_df.head()
```

```python
# calculating wickets

delivery_df["wickets_left"]=10 - delivery_df.groupby("match_id").cumsum()["player_dismissed"]
```

```python
delivery_df.head()
```

```python
# current run rate

delivery_df['cur_run_rate'] = (delivery_df['current_score']*6)/(120-delivery_df['balls_left'])

# required run rate

delivery_df['req_run_rate'] = (delivery_df['runs_left']*6)/(delivery_df['balls_left'])
```

```python
delivery_df["result"]=np.where(delivery_df["batting_team"]==delivery_df["winner"],1,0)
```

```python
delivery_df.head()
```

```python
final_df = delivery_df[['batting_team','bowling_team','city','runs_left',
                        'balls_left','wickets_left','target_runs','cur_run_rate',
                        'req_run_rate','result']]

final_df.head()
```

### Splitting the data

```python
data=final_df.copy()
test = data['result']
train = data.drop(['result'],axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(train,test,test_size=0.2,random_state=1)
```

### Model Building

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn import metrics
```

```python
cf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
],remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',cf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
print(metrics.accuracy_score(y_test,y_pred))
```
